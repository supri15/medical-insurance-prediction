```python
from flask import Flask, render_template, request
import pandas as pd
import pickle

# Initialize the Flask application
app = Flask(__name__)

# Load and preprocess the data
data = pd.read_csv("insurance.csv")

# Encode categorical variables
data['sex'] = data['sex'].map({'female': 0, 'male': 1})
data['smoker'] = data['smoker'].map({'yes': 1, 'no': 0})
data['region'] = data['region'].map({'southwest': 1, 'southeast': 2,
                                      'northwest': 3, 'northeast': 4})

# Prepare features and target variable
X = data.drop(['charges'], axis=1)
y = data['charges']


with open('random_forest_model.pkl', 'rb') as f:
    rf = pickle.load(f)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get input values from the form
        inputs = [float(request.form['age']),
                  request.form['sex'],
                  float(request.form['bmi']),
                  int(request.form['children']),
                  request.form['smoker'],
                  request.form['region']]

        # Map categorical values to integers
        inputs[1] = 0 if inputs[1] == 'female' else 1
        inputs[4] = 1 if inputs[4] == 'yes' else 0
        inputs[5] = {'southwest': 1, 'southeast': 2, 'northwest': 3, 'northeast': 4}[inputs[5]]

        # Convert inputs to a DataFrame
        input_df = pd.DataFrame([inputs], columns=['age', 'sex', 'bmi', 'children', 'smoker', 'region'])

        # Predict using all models
        y_pred3 = rf.predict(input_df)[0]

        return render_template('index.html', prediction_rf=y_pred3)
    except Exception as e:
        return str(e)

if __name__ == '__main__':
    app.run(port=5001)
```
